```python
global username
global cnic
global _username
global _password
global admin_username
global admin_pass

global bill_0
global bill1
global bill2
global bill3
global name
global destination
global quantity
global numofitems
items = []
global totalbill

global reserve_name
global date
global seats
global time


def mainscreen():
    print("\n\t WELCOME TO AIRLINE MANAGEMENT SYSTEM ")
    print(
        "\n\tIf you are our VIP Customer. Press 1.\n\tIf you are our New Customer. Press 2 for Registration.\n\tPress 3 for Admin login ")
    choice = int(input("\tEnter Selected Option:"))
    if choice == 1:
        vip_customer_login()
    elif choice == 2:
        customerregister()
    elif choice == 3:
        adminlogin()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()


# ----------------------------------------------------------------------------------------------------------

def vip_customer_login():
    print("\n")
    print("\n")
    print("\t\t-------IF you are our VIP Customer. login below-------\n")
    _username = str(input("\tEnter Username:"))
    _password = str(input("\tEnter Password:"))
    if not _username or not _password:
        print("!!!!_Fields cannot be empty._!!!!")
        vip_customer_login()
    else:
        with open("registrations.txt") as f:
            if _username in f.read():
                with open("registrations.txt") as fp:
                    if _password in fp.read():
                        print("\tWelcome ," + _username)
                        print("----- WELCOME TO OUR VIP PANEL -----\n Tell us what can we do for you ?")
                        print("\tPress 1 for a pre booking.\n\tPress 2 for Normal booking.\n\tPress 3 to Logout")
                        choice = int(input("Enter Option = \t"))
                        if choice == 1:
                            pre_booking()
                        elif choice == 2:
                            normal_bookings()
                        elif choice == 3:
                            mainscreen()
                        else:
                            print("\t!!!!_Enter correct option_!!!!\n")
                            print("\tWelcome ," + _username)
                            print("----- WELCOME TO OUR VIP PANEL -----\n Tell us what can we do for you ?")
                            print("\tPress 1 for a pre booking.\n\tPress 2 for Normal booking.\n\tPress 3 to Logout")
                            choice = int(input("\t"))
                            if choice == 1:
                                pre_booking()
                            elif choice == 2:
                                normal_bookings()
                            elif choice == 3:
                                mainscreen()
                            else:
                                print("\t!!!!_Enter correct option_!!!!\n")

                    else:
                        print("\t!!Wrong Password.!!")
                        vip_customer_login()

            else:
                print("\t!!!_Wrong Username._!!!\n\tYou need to register first!")
                vip_customer_login()


def customerregister():
    print("\n")
    print("\n")
    print("\t\t\t\t Customer Register Screen \n")
    username = str(input("\tEnter your Name  = "))
    cnic = str(input("\tEnter your CNIC Number = "))
    if not username or not cnic:
        print("!!!!_Fields cannot be empty._!!!!")
        customerregister()
    else:
        file = open('registrations.txt', 'a')
        file.write(username + "," + cnic + "\n")
        print(("\t===Registered Sucessfully==="))
        file.close()
        mainscreen()


def pre_booking():
    items = []
    global totalbill
    print("\n\t")
    print("\t\t------- PRE BOOKING LIST -------")
    print("\t PRE BOOKING IS AVAILABLE ONLY FOR THE CAPITAL CITIES OF EACH PROVINCE")
    print("\t---------------\t\t--------------\t\t\t------")
    print("\t Karachi - Rs: 10,000 Rs 200\t\tLahore - Rs: 50,000\t\t\tPeshawar - Rs: 100,000")
    print("\tQuetta - Rs: 10,000\tKashmir - Rs: 500,000\n")
    print()
    name = str(input("Enter Your Name: "))
    number_of_tickets = int(input("For how many persons would you like to Pre-book on your VIP Card.(Only 2 Tickets)"))
    if number_of_tickets == 1:
        print("Pre- Booking Options: \n"
              "1. Karachi\n"
              "2. Lahore\n"
              "3. Peshawar\n"
              "4. Quetta\n"
              "5. Kashmir.")
        destination = str(input("Where do you wanna go?"))
        if destination == "Karachi" or destination == "karachi":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 10000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Lahore" or destination == "lahore":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 50000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Peshawar" or destination == "peshawar":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 100000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Quetta" or destination == "quetta":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 10000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Kashmir" or destination == "kashmir":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 500000 * quantity1
            print(f"Bill: {bill_0}")

        else:
            print("Flight Unavailable.!")

        print("Would you like to book Tickets for another flight ?(VIP only)")
        choice = str(input("Enter Y for Yes N for No:"))

        if choice == "y" or choice == "Y":
            destination = str(input("Where do you wanna go?"))
            if destination == "Karachi" or destination == "Karachi":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 10000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Lahore" or destination == "Lahore":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 50000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Peshawar" or destination == "Peshawar":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 100000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Quetta" or destination == "Quetta":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 10000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Kashmir" or destination == "Kashmir":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 500000 * quantity1
                print(f"Bill: {bill_0}")
            print("\Tickets Booked")
            file = open("orders.txt", 'a')
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(
                numofitems) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(totalbill) + "\n")
            with open('orders.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            file = open("allbooking.txt.txt", 'a')
            file.write("Pre Booking"+"\n")
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(
                numofitems) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(totalbill) + "\n")
            with open('allbooking.txt.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            print("------------------------------------------------")
            print("\t**Billing**")
            print("\t===================")
            print("Name:" + name)
            print("Booked 1 Tickets")
            print("====================\n")
            print("Total Cost = ", bill_0)
            print("------------------------------")
            print("Totalbill    \t\t", totalbill)
            print("------------------------------------------------")
            print("\n")
            print("\tPress 1 for another Pre-booking\n\tPress 2 for a Normal Booking\n\tPress 3 to Logout")
            choice = int(input("\t"))
            if choice == 1:
                pre_booking()
            elif choice == 2:
                normal_bookings()
            elif choice == 3:
                mainscreen()
            else:
                print("!!!Enter correct Input_!!!")
                print("\tPress 1 for another Pre-booking\n\tPress 2 for a Normal Booking\n\tPress 3 to Logout")
                choice = int(input("\t"))
                if choice == 1:
                    pre_booking()
                elif choice == 2:
                    normal_bookings()
                elif choice == 3:
                    mainscreen()



        else:
            print("\tOrder Placed")
            file = open("orders.txt", 'a')
            file.write("-->Items, Name of Order-->" + name + ", Number of Tickets booked-->" + str(quantity1) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(bill_0) + "\n")
            with open('orders.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            file = open("allbooking.txt", 'a')
            file.write("Pre Booking" + "\n")
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(quantity1) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(bill_0) + "\n")
            with open('allbooking.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            print("------------------------------------------------")
            print("\t**Billing**")
            print("\t===================")
            print("Name:" + name)
            print("Ordered 1 item")
            print("1st item:", items, sep="\n")
            print("Price of 1st item:", bill_0)
            print("------------------------------")
            print("Totalbill    \t\t", bill_0)
            print("------------------------------------------------")
            print("\n")
            print("\Press 3 to Logout")
            choice = int(input("\t"))
            if choice == 3:
                mainscreen()
            else:
                print("!!!Enter correct Input_!!!")
                print("\tPress 1 for another Pre-Order\n\tPress 2 to make a Normal reservation\n\tPress 3 to Logout")
                choice = int(input("\t"))
                if choice == 3:
                    mainscreen()

    elif number_of_tickets == 2:
        print("Pre- Booking Options: \n"
              "1. Karachi\n"
              "2. Lahore\n"
              "3. Peshawar\n"
              "4. Quetta\n"
              "5. Kashmir.")
        destination = str(input("Where do you wanna go?"))
        if destination == "Karachi" or destination == "Karachi":
            items.append(destination)
            quantity1= int(input("How many Tickets would you like to register?"))
            bill_0 = 2 * 10000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Lahore" or destination == "Lahore":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 2 * 50000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Peshawar" or destination == "Peshawar":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 2 * 100000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Quetta" or destination == "Quetta":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 2 * 10000 * quantity1
            print(f"Bill: {bill_0}")
        elif destination == "Kashmir" or destination == "Kashmir":
            items.append(destination)
            quantity1 = int(input("How many Tickets would you like to register?"))
            bill_0 = 2 * 500000 * quantity1
            print(f"Bill: {bill_0}")

        else:
            print("Flight Unavailable.!")

        print("Would you like to book Tickets for another flight ?(VIP only)")
        choice = str(input("Enter Y for Yes N for No:"))

        if choice == "y" or choice == "Y":
            destination = str(input("Where do you wanna go?"))
            if destination == "Karachi" or destination == "Karachi":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 2 * 10000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Lahore" or destination == "Lahore":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 2 * 50000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Peshawar" or destination == "Peshawar":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 2 * 100000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Quetta" or destination == "Quetta":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 2 * 10000 * quantity1
                print(f"Bill: {bill_0}")
            elif destination == "Kashmir" or destination == "Kashmir":
                items.append(destination)
                quantity1 = int(input("How many Tickets would you like to register?"))
                bill_0 = 2 * 500000 * quantity1
                print(f"Bill: {bill_0}")
            print("\Tickets Booked")
            file = open("orders.txt", 'a')
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(
                quantity1) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + "\n")
            with open('orders.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            file = open("allbooking.txt", 'a')
            file.write("Pre Booking" + "\n")
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(
                quantity1) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + "\n")
            with open('allbooking.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            print("------------------------------------------------")
            print("\t**Billing**")
            print("\t===================")
            print("Name:" + name)
            print("Booked Tickets")
            print("====================\n")
            print("Total Cost = ", bill_0)
            print("------------------------------")
            print("------------------------------------------------")
            print("\n")
            print("\tPress 1 for another Pre-booking\n\tPress 2 for a Normal Booking\n\tPress 3 to Logout")
            choice = int(input("\t"))
            if choice == 1:
                pre_booking()
            elif choice == 2:
                normal_bookings()
            elif choice == 3:
                mainscreen()
            else:
                print("!!!Enter correct Input_!!!")
                print("\tPress 1 for another Pre-booking\n\tPress 2 for a Normal Booking\n\tPress 3 to Logout")
                choice = int(input("\t"))
                if choice == 1:
                    pre_booking()
                elif choice == 2:
                    normal_bookings()
                elif choice == 3:
                    mainscreen()



        else:
            print("\tOrder Placed")
            file = open("orders.txt", 'a')
            file.write("-->Items, Name of Order-->" + name + ", Number of Tickets booked-->" + str(
                numofitems) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(bill_0) + "\n")
            with open('orders.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            file = open("allbooking.txt.txt", 'a') 
            file.write("Pre Booking" + "\n")
            file.write("-->Pre - bookings, Name of Booking-->" + name + ", Number of Tickets-->" + str(
                numofitems) + ", Price of 1st item-->" + str(bill_0) + ", Bill-->" + str(totalbill) + "\n")
            with open('allbooking.txt.txt', 'a+') as filehandle:
                for listitem in items:
                    filehandle.write('%s ,' % listitem)
            file.write("\n")
            file.close()

            print("------------------------------------------------")
            print("\t**Billing**")
            print("\t===================")
            print("Name:" + name)
            print("Ordered 1 item")
            print("1st item:", items, sep="\n")
            print("Price of 1st item:", bill_0)
            print("------------------------------")
            print("Totalbill    \t\t", bill_0)
            print("------------------------------------------------")
            print("\n")
            print("Press 3 to Logout")
            choice = int(input("\t"))
            if choice == 3:
                mainscreen()
            else:
                print("!!!Enter correct Input_!!!")
                print("Press 3 to Logout")
                choice = int(input("\t"))
                if choice == 3:
                    mainscreen()


def normal_bookings():
    global reserve_name
    global date
    global seats
    global time
    print("\n\t=====================WElCOME TO NORMAL BOOKING========================")
    print("Ticket Bookings available for Cities:\n"
          "Karachi, Hyderabad, Lahore, Peshawar, Quetta, Multan, Gilgit, Islamabad, Rawalpindi,"
          "Faislabad, Gujranwala, Sialkot"
          "Sukkur,Larkana, Gawadar")

    print("\t\t------- Book your Ticket below. -------")
    reserve_name = str(input("\n Enter name for booking:"))
    if not reserve_name:
        print("\t!!!Enter a name!!!")
        normal_bookings()
    else:
        seats = int(input("How many tickets do you want to register on this CNIC ?"))
        date = str(input("\tEnter date for booking in DD/MM/YY format:"))
        time = str(input("\tEnter time for booking in format XX.XX PM:"))

        print("Enter the Class in which you'd like to travel")
        air_class = input("Press A for a First Class Ticket. Press B for an Economy Class Ticket.")

        if air_class == "A" or air_class == "a":
            extra_amount = 50000
            number_of_tickets = int(input("For how many tickets would you like to book ?"))

            destination = str(input("Where do you wanna go?"))
            if destination == "Karachi" or destination == "karachi" or destination == "Lahore" or destination == "lahore" or destination == "Peshawar" or destination == "peshawar" or destination == "Quetta" or destination == "quetta" or destination == "Islamabad" or destination == "islamabad":
                items.append(destination)
                bill_0 = extra_amount * 20000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "Multan" or destination == "multan" or destination == "Rawalpindi" or destination == "rawalpindi" or destination == "Hyderabad" or destination == "hyderabad" or destination == "Sialkot" or destination == "sialkot":
                items.append(destination)
                bill_0 = extra_amount * 50000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination ==  destination == "gilgit" or destination == "Gilgit" or destination == "larkana" or destination == "Larkana" or destination == "gwadar" or destination == "Gwadar":
                items.append(destination)
                bill_0 = extra_amount * 100000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "abbotabad" or destination == "charsadda" or destination == "Gawadar" or destination == "jhelum":
                items.append(destination)
                bill_0 = extra_amount * 70000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "sawat" or destination == "Khairpur" or destination == "Khairpur" or destination == "Chakwal" or destination == "chakwal" or destination == "Kashmir":
                items.append(destination)
                bill_0 = extra_amount * 120000 * number_of_tickets
                print(f"Bill: {bill_0}")

            else:
                print("Enter a Valid City.!")

        elif air_class == "B" or air_class == "b":
            extra_amount = 10000

            number_of_tickets = int(input("For how many tickets would you like to book ?"))

            destination = str(input("Where do you wanna go?"))
            if destination == "Karachi" or destination == "karachi" or destination == "Lahore" or destination == "lahore" or destination == "Peshawar" or destination == "peshawar" or destination == "Quetta" or destination == "quetta" or destination == "Islamabad" or destination == "islamabad":
                items.append(destination)
                bill_0 = extra_amount * 20000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "Multan" or destination == "multan" or destination == "Rawalpindi" or destination == "rawalpindi" or destination == "Hyderabad" or destination == "hyderabad" or destination == "Sialkot" or destination == "sialkot":
                items.append(destination)
                bill_0 = extra_amount * 50000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == destination == "gilgit" or destination == "Gilgit" or destination == "larkana" or destination == "Larkana" or destination == "gwadar" or destination == "Gwadar":
                items.append(destination)
                bill_0 = extra_amount * 100000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "abbotabad" or destination == "charsadda" or destination == "Gawadar" or destination == "jhelum":
                items.append(destination)
                bill_0 = extra_amount * 70000 * number_of_tickets
                print(f"Bill: {bill_0}")
            elif destination == "sawat" or destination == "Khairpur" or destination == "Khairpur" or destination == "Chakwal" or destination == "chakwal" or destination == "Kashmir":
                items.append(destination)
                bill_0 = extra_amount * 120000 * number_of_tickets
                print(f"Bill: {bill_0}")

            else:
                print("Enter a Valid City.!")

        print("\n")
        file = open("normal_bookings.txt", 'a')
        file.write(str(seats) + "-->Number of seats, " + reserve_name + "-->Name , " + str(
            date) + "--> Date of Booking, " + str(time) + "-->Time of Booking" + "\n")
        file.close()
        file = open("allbooking.txt", 'a') 
        file.write("Normal Booking"+"\n")
        file.write(str(seats) + "-->Number of seats, " + reserve_name + "-->Name , " + str(
            date) + "--> Date of Booking, " + str(time) + "-->Time of Booking" + "\n")
        file.close()
        print("\t=======Reservation Done=======")
        print("\t********")
        print("\t------Details-------")
        print("\tName by which reservation is done:" + reserve_name)
        print("\tReserved for " + str(seats) + " people")
        print("\tTime for which reserved is:" + time)
        print("\tDate for which reserved is:" + date)
        print(f"\tBill: {bill_0}")
        print("\t********")
        print("\tPress 1 for Pre booking\n\tPress 2 for Normal booking\n\tPress 3 to Logout")
        choice = int(input("\t"))
        if choice == 1:
            pre_booking()
        elif choice == 2:
            normal_bookings()
        elif choice == 3:
            mainscreen()
        else:
            print("!!!Enter correct Input_!!!")
            print("\tPress 1 for Pre booking\n\tPress 2 for Normal booking\n\tPress 3 to Logout")
            choice = int(input("\t"))
            if choice == 1:
                pre_booking()
            elif choice == 2:
                normal_bookings()
            elif choice == 3:
                mainscreen()

def adminlogin():
    print("\n")
    print("\n\t ------------------------- WELCOME TO THE ADMIN SCREEN -------------------------")
    print("\t\t\t\t\t-------Admin login screen-------")
    admin_username = str(input("\tEnter Username:"))
    admin_pass = str(input("\tEnter Password:"))
    if admin_username == "Noor" or admin_username == "noor" or admin_username == "Talha" or admin_username == "talha":
        if admin_pass == "admin":
            print("\t\tLogin Successful.\n\t\t=====Welcome Admin : " + admin_username + "=====")
            admindashboard()
        else:
            print("\tIncorrect Password. Try again.!\n")
            adminlogin()
    else:
        print("\tInvalid Username. Try again.!\n")
        adminlogin()


def admindashboard():
    print("\n")
    print("\n\t===================== WELCOME TO THE ADMIN DASHBOARD ========================")
    print("\t\t\t\t\t-------Hello Admin-------")
    print(
        "\n\t Press 1 to view CUstomers Log\n\tPress 2 to view Ticket Resgistrations\n\tPress 3 to sort Registrations in Order\n\tPress 4 to view recommendation\n\tPress 5 to search for a Reservation\n\tPress 6 view name of Employees\n\tPress 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        view_orders()
    elif choice == 2:
        view_reservations()
    elif choice == 3:
        sort_reservations()
    elif choice == 4:
        viewrecommendation()
    elif choice == 5:
        searching_reservations()
    elif choice == 6:
        employeenames()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()


def view_orders():
    print(
        "*************************")
    print("\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\tOrder Details")
    file = open("orders.txt", 'r')
    _orders = file.read()
    print(_orders)
    file.close()
    file = open("allbooking.txt", 'r')
    __orders = file.read()
    print(__orders)
    file.close()
    print(
        "************************")
    print("\n")
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice==1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()


def view_reservations():
    print("*************")
    print("\t\t\t\t\t\t\t\t\tReservation Details")
    file = open("reservations.txt", 'r')
    _reservations = file.read()
    print(_reservations)
    file.close()

    file = open("allbooking.txt.txt", 'r') 
    __reservations = file.read()
    print(__reservations)
    file.close()

    print("*************")
    print("\n")
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()


def sort_reservations():
    print(("\n\tPress B for bubble sort\n\tPress S for Selection sort"))
    choice = str(input())
    if choice == "b" or choice == "B":
        bubblesort()
    elif choice == "s" or choice == "S":
        selectionsort()
    else:
        print("Enter correct input")
        sort_reservations()



def bubblesort():
    input_file = open('reservations.txt')
    input__file = open('allbooking.txt.txt')

    lst = []
    for line in input_file:
        lst.append(line.strip())

    fst = []    #by me
    for line in input__file:
        fst.append(line.strip())

    def bubble_sort(arr):
        n = len(arr)

        for i in range(n):

            for j in range(0, n - i - 1):

                if arr[j] > arr[j + 1]:
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]
        return arr

    print("\t**Applying Bubble sort.....Done**")
    print(*bubble_sort(lst), sep="\n")
    print(*bubble_sort(fst), sep="\n")
    file = open("allbooking.txt.txt", 'r') 
    __reservations = file.read()
    print(__reservations)
    file.close()
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()


def selectionsort():
    input_file = open('reservations.txt')
    input__file = open('allbooking.txt.txt')

    lst = []
    for line in input_file:
        lst.append(line.strip())

    fst = []    #by me
    for line in input__file:
        fst.append(line.strip())

    def selection_sort(arr):

        for i in range(len(arr)):

            min_idx = i
            for j in range(i + 1, len(arr)):
                if arr[min_idx] > arr[j]:
                    min_idx = j

            arr[i], arr[min_idx] = arr[min_idx], arr[i]
        return arr

    print("\t**Applying Selection sort.....Done**")
    print(*selection_sort(lst), sep="\n")
    print(*selection_sort(fst), sep="\n")
    file = open("allbooking.txt.txt", 'r') 
    __reservations = file.read()
    print(__reservations)
    file.close()
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        mainscreen()



def searching_reservations():
    global datatosearch

    with open("allbooking.txt.txt") as openfile:
        datatosearch = input("Enter name of reservation:")
        for line in openfile:
            for part in line.rsplit("\n"):
                if datatosearch in part:
                    print("\tData Found")
                    print(part)
                    print("\n\tPress 7 to go back to main menu ")
                    choice = int(input("\t"))
                    if choice == 7:
                        mainscreen()
                    else:
                        print("\t!!!!_Enter correct option_!!!!\n")
                        mainscreen()

                else:
                    print("Data not found")
                    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
                    choice = int(input("\t"))
                    if choice == 1:
                        admindashboard()
                    elif choice == 7:
                        mainscreen()
                    else:
                        print("\t!!!!_Enter correct option_!!!!\n")
                        mainscreen()


def searchorder():
    input_file = open('reservations.txt')
    lst = input_file.read()
    for line in input_file:
        lst.append(line.strip())
    global datatosearch
    datatosearch = input("\tEnter name of reservation:")

    def binary_search(arr, low, high, x):
        # Check base case
        if high >= low:

            mid = (high + low) // 2

            # If element is present at the middle itself
            if arr[mid] == x:
                return mid

            # If element is smaller than mid, then it can only
            # be present in left subarray
            elif arr[mid] > x:
                return binary_search(arr, low, mid - 1, x)

            # Else the element can only be present in right subarray
            else:
                return binary_search(arr, mid + 1, high, x)

        else:
            # Element is not present in the array
            return -1

    result = binary_search(lst, 0, len(lst) - 1, datatosearch)
    if (result == -1):
        print("Not Found")
    else:
        print("\tData found \n" + str(result))


def searching_order():
    global datatosearch
    with open("orders.txt") as openfile:
        datatosearch = input("Enter name of booking:")
        for line in openfile:
            for part in line.rsplit("\n"):
                if datatosearch in part:
                    print("\tData Found")
                    print(part)
                    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
                    choice = int(input("\t"))
                    if choice == 1:
                        admindashboard()
                    elif choice == 7:
                        mainscreen()
                    else:
                        print("\t!!!!_Enter correct option_!!!!\n")
                        mainscreen()

                else:
                    print("Data not found")
                    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
                    choice = int(input("\t"))
                    if choice == 1:
                        admindashboard()
                    elif choice == 7:
                        mainscreen()
                    else:
                        print("\t!!!!_Enter correct option_!!!!\n")
                        mainscreen()


class Queue:
    def init(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)


def viewrecommendation():
    q = Queue()
    q.enqueue("Customers")
    q.enqueue("Employees")
    q.enqueue("Assets")
    q.enqueue("Partners")
    q.enqueue("Branches")

    print("Queue Size: "+str(q.size()))
    print("First new item we are going to include in our menu is: ", q.dequeue())
    print("Second new item we are going to include in our menu is: ", q.dequeue())
    print("Third new item we are going to include in our menu is: ", q.dequeue())
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        admindashboard()


class Node:
    def init(self, data):
        self.data = data
        self.next = None
        self.prev = None


class Stack:
    def init(self):
        self.head = None

    def push(self, data):

        if self.head is None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            self.head.prev = new_node
            new_node.next = self.head
            new_node.prev = None
            self.head = new_node

    def pop(self):
        if self.head is None:
            return None
        else:
            temp = self.head.data
            self.head = self.head.next
            self.head.prev = None
            return temp

    def top(self):

        return self.head.data

    def length(self):
        temp = self.head
        count = 0
        while temp is not None:
            count = count + 1
            temp = temp.next
        return count

    def rem_printingofstack(self):

        temp = self.head
        while temp is not None:
            print(str(temp.data) + " ")
            temp = temp.next


def employeenames():
    stack = Stack()
    stack.push("Malik")
    stack.push("Afnan")
    stack.push("Dani")
    stack.push("Dilshad")
    stack.push("Anas")
    stack.push("Safi")
    stack.push("Talha")
    stack.push("Haris")
    stack.push("Jameel")
    stack.push("Asad")

    print("\t\tEmployess present in stack are:")
    print("\t\t------------------------------")
    stack.rem_printingofstack()
    print("\t\t-----------------------------------------------------------")
    print("\t\tNumber of Employees present in Stack:" + str(stack.length()))
    print("\t\tNewest Employee is: ", stack.top(), " (top name in stack)")
    print("\t\tMost newest employee is:", stack.pop(), " (first name popped out of stack)")
    print("\t\tSecond newest employee is:", stack.pop(), " second name popped out of stack")
    print("\t\t-----------------------------------------------------------")
    print("\n\t\tStack after popping of two times is:")
    print("\t\t---------------------------------------")
    stack.rem_printingofstack()
    print("\n\tPress 1 For Admin Actions And 7 to go back to main menu ")
    choice = int(input("\t"))
    if choice == 1:
        admindashboard()
    elif choice == 7:
        mainscreen()
    else:
        print("\t!!!!_Enter correct option_!!!!\n")
        admindashboard()


# ___Secondary Part___
class Node:
    def init(self, data=None):
        self.data = data
        self.next = None


# Linked List
class Airline_Management_System:
    def init(self):
        self.head = None
        self.tail = None

    # DS queue
    def append_value(self, x):
        if not isinstance(x, Node):
            x = Node(x)
        if self.head is None:
            self.head = x
        else:
            self.tail.next = x
        self.tail = x

    def length(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    # DS
    def insertion_sort(self):
        current = self.head
        for _ in range(self.length() - 1):
            temp = current.next
            while temp:
                if temp.data < current.data:
                    current.data, temp.data = temp.data, current.data
                temp = temp.next
            current = current.next
        return self

    # DS
    def remove_value(self, key):
        current = self.head
        if current:
            if current.data == key:
                self.head = current.next
                return
        while current:
            if current.data == key:
                break
            previous = current
            current = current.next
        if current is None:
            return
        previous.next = current.next

    def search_by_index(self, i):
        current = self.head
        count = 0
        while current:
            if i == count:
                return current.data
            current = current.next
            count += 1

    # DS
    def binary_search(self, key):
        start = 0
        stop = self.length() - 1
        self.insertion_sort()
        while start <= stop:
            mid = (start + stop) // 2
            if key == self.search_by_index(mid):
                return f'{key} found at index: {mid}'
            elif key > self.search_by_index(mid):
                start = mid + 1
            else:
                stop = mid - 1
        return f'{key} not found'

    def str(self):
        to_print = ''
        current = self.head
        while current is not None:
            to_print += f'{current.data}->'
            current = current.next
        if to_print:
            return f'[{to_print[:-2]}]'
        return '[]'

    def file(self, something):
        with open('filename.txt', 'a') as file:
            file.write(f'{something}\n')

    def Airport_policies(self):
        print()
        print("\t\t-------------- OUR AIRPORT POLICIES ARE: --------------\n")
        f = open("policy.txt", "r")
        print(f.read())
        print()



mainscreen()
```
